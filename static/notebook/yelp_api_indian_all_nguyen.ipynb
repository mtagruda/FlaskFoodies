```python
# American restaurants in Houston
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
from yelpapi import YelpAPI
from config import yelp_API
```

```python
url='https://api.yelp.com/v3/businesses/search'
headers = {'Authorization':'Bearer '+ yelp_API}
```

```python
# Indian restaurants in Houston

df_yelp_houston_indian=pd.DataFrame()

offset= [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950]

for item in offset:
    params_set1 = {'location': 'Houston, TX','limit': 50,'offset': item,'alias': 'indian','term': 'restaurant'}
    response = requests.get(url, headers=headers, params=params_set1)
    response_json=response.json()
    print(response_json)
    res_lat=[]
    res_long=[]
    res_name=[]
    res_rating=[]
    res_vote=[]
    for respo in response_json:
        try:
            nam=response_json['businesses']
        except:
            print('NA')
        for na in nam:
            try:
                name= na['name']
                rating= na['rating']
                vote=na['review_count']
                lat=na['coordinates']['latitude']
                long=na['coordinates']['longitude']
                res_name.append(name)
                res_rating.append(rating)
                res_vote.append(vote)
                res_lat.append(lat)
                res_long.append(long)
            except:
                print('NA')
    df_yelp=pd.DataFrame({'Name': res_name,'Latitude':res_lat,'Longitude': res_long,'Votes': res_vote,'Rating': res_rating})
    df_yelp_houston_indian=df_yelp_houston_indian.append(df_yelp)

df_yelp_houston_indian['Type']='indian'
df_yelp_houston_indian['City']='houston'
```

```python
df_yelp_houston_indian
```

```python
# Drop duplicates:
df_yelp_houston_indian_all=df_yelp_houston_indian.drop_duplicates(keep='first')
df_yelp_houston_indian_all
```

```python
#### Save raw data to csv
df_yelp_houston_indian_all.to_csv('C:\workspace\Resources\df_yelp_houston_indian_all_NN.csv',index=False)
```
